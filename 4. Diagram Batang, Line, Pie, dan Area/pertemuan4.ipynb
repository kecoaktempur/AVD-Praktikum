```python
# Import package
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib as mpl
```

```python
# Import dataset
dataset = pd.read_csv(r'D:\Coolyeah\Mata Kuliah\SMT 5\Analisis Visualisasi Data Praktikum\4. Diagram Batang, Line, Pie, dan Area\McDonaldsMenuNutritionV2.csv')
dataset.head(10)
```

```python
datadrop = dataset.drop(columns=[
    'Item', 'Calories'
    ])
datadrop.head(10)
```

```python
datasum = datadrop.sum()
datasum.head(12)
```

```python
# Diagram Batang

df = pd.DataFrame({'Weight Watchers Pnts':datasum.index, 'Val':datasum})
ax = df.plot.bar(x='Weight Watchers Pnts', y='Val', rot=0, figsize=(8,4))
```

```python
# Diagram Garis

plt.plot(datasum, color='red')
plt.ylabel('Value')
plt.title('Visualisasi menggunakan diagram garis')
plt.show()
```

```python
# Diagram Pie

plt.figure()
labels = ['Calories from Fat', 'Total Fat (g)', 'Saturated Fat (g)', 'Trans Fat (g)', 'Cholesterol (mg)', 'Sodium (mg)', 'Carbs (g)', 'Fiber (g)', 'Sugars (g)', 'Protein (g)', 'Weight Watchers Pnts']
plt.pie(datasum, labels=labels, autopct='%.2f')
plt.title('Visualisasi menggunakan diagram pie')
plt.show()
```

```python
# Area Plots

dataset.plot(kind='area')

plt.title('Area Plots')
plt.ylabel('Calories')

plt.show()
```
