```python
# Import Package dan Library
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
```

```python
# Import Dataset
data_1 = pd.read_csv(r"D:\Coolyeah\Mata Kuliah\SMT 5\Analisis Visualisasi Data Praktikum\5. High Dimensional dan Multivariate\shopping_data.csv")

col_1 = data_1['32']
col_2 = data_1['15']
col_3 = data_1['39']

df_3d = pd.DataFrame([col_1.head(5), col_2.head(5), col_3.head(5)]).transpose()
df_3d.head(5)
```

1. Diagram Batang


```python
colors_1 = ['r', 'b', 'g', 'y', 'b', 'w']
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
z = list(df_3d)

for n, i in enumerate(df_3d):
    print('n', n)
    xs = np.arange(len(df_3d[i]))
    ys = [float(val) for val in df_3d[i]]
    zs = z[n]

    cs = colors_1[n]
    print('xs:', xs, 'ys:', ys, 'zs:', zs, 'cs:', cs)
    ax.bar(xs, ys, zs, zdir='y', color=cs, alpha=0.8)

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

plt.show()
```

2. Diagram Batang Multiple Dataframes

```python
columns = ['A', 'B', 'C']
df_name = ['Age', 'Annual Income', 'Spending Score']
colors_3 = ('r', 'g', 'b')

main_df = [df_3d]

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

plt.xticks([i for i in range(len(columns))], columns)
plt.yticks([i for i in range(len(df_name))], df_name)

xs = []
for i in range(len(main_df)):
    for j in range(len(columns)):
        xs.append(i + j * 0.1)

for c1, color in enumerate(colors_3):
    x = data_1['Age'].head(4)
    y = data_1['Annual Income (k$)'].head(3)
    z = data_1['Spending Score (1-100)'].head(3)

    ax.bar(xs, y, z, zdir='y', alpha=0.8, color=colors_3)

plt.title('Multiple DataFrames')
plt.show()
```

3. ScatterPlot

```python
data = [col_1, col_2, col_3]

colors_3 = ('r', 'g', 'b')
groups = ("Age", "Annual Income", "Spending Score")

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

for data, color, group in zip(data, colors_3, columns):
    x = data_1['Age']
    y = data_1['Annual Income (k$)']
    z = data_1['Spending Score (1-100)']
    ax.scatter(x, y, z, alpha=0.8, color=color,
               edgecolors='none', s=30, label=group)

plt.title('3D Scatterplot')
plt.legend(loc=2)
plt.show()
```
